### 1. ¿Qué ave tiene el menor número de vértebras en el cuello?
Para esto, cargamos bird-neck-bones.csv y buscamos el valor mínimo en la columna de vértebras.

```python
import pandas as pd

# Cargar y limpiar nombres de columnas
birds_df = pd.read_csv('bird-neck-bones.csv')
birds_df.columns = birds_df.columns.str.strip()

# Encontrar la fila con el mínimo de vértebras (usando la columna 'species')
min_row = birds_df.loc[birds_df['neck_vertebrae'].idxmin()]

print(f"El ave con menos vértebras es: {min_row['species']}")
print(f"Cantidad de vértebras: {min_row['neck_vertebrae']}")
```

### 2. ¿Tienen los humanos más huesos en los brazos o en las piernas?
Utilizaremos adult-human-skeleton.csv y filtraremos por la columna region.

```python
import pandas as pd

# Cargar y limpiar nombres de columnas
human_df = pd.read_csv('adult-human-skeleton.csv')
human_df.columns = human_df.columns.str.strip()

# Filtrar y contar filas por región
arm_bones = human_df[human_df['region'] == 'arm'].shape[0]
leg_bones = human_df[human_df['region'] == 'leg'].shape[0]

print(f"Huesos en los brazos (región 'arm'): {arm_bones}")
print(f"Huesos en las piernas (región 'leg'): {leg_bones}")

if leg_bones > arm_bones:
    print("Los humanos tienen más huesos registrados en las piernas.")
```

### 3. ¿Cuántas costillas tienen los humanos?
Utilizaremos el método .str.contains() para localizar las filas que mencionan "rib" (costilla) y sumaremos su cantidad.

```python
import pandas as pd

# Cargar y limpiar (si no se hizo antes)
human_df = pd.read_csv('adult-human-skeleton.csv')
human_df.columns = human_df.columns.str.strip()

# Filtrar filas donde el nombre contiene "rib"
ribs_df = human_df[human_df['name'].str.contains("rib", case=False)]

# Contar el número de filas (cada fila es una costilla)
total_ribs = ribs_df.shape[0]

print(f"Número de filas encontradas con 'rib': {total_ribs}")
print(f"El total de costillas es: {total_ribs}")
```
